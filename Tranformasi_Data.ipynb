#  Mengatasi Data Outlier

```python
import pandas as pd
import scipy.stats as stats


columns = ['Heart Rate', 'Sleep Duration']

for column in columns:
    df[column] = stats.mstats.winsorize(df[column], limits=[0.05, 0.05])

print(df[['Heart Rate', 'Sleep Duration']].describe())

df.to_csv('data_winsorized.csv', index=False)
```

dengan menerapkan Winsorizing pada kolom Heart Rate dan Sleep Duration untuk menangani outliers dengan mengganti nilai ekstrem (paling kecil dan paling besar) dengan nilai persentil tertentu yang lebih rendah atau lebih tinggi. Dalam kasus ini, kamu menggunakan batas persentil 5% dan 95%, artinya nilai yang berada di bawah persentil 5% akan diganti dengan nilai di persentil 5%, dan nilai yang di atas persentil 95% akan diganti dengan nilai di persentil 95%.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# Heart Rate
plt.subplot(1, 2, 1)
plt.boxplot(df['Heart Rate'])
plt.title('Heart Rate')

# Sleep Duration
plt.subplot(1, 2, 2)
plt.boxplot(df['Sleep Duration'])
plt.title('Sleep Duration')

plt.show()
```

kode ini bertujuan untuk membuat boxplot guna memeriksa outlier pada dua variabel, yaitu Heart Rate dan Sleep Duration, sebelum dan setelah proses Winsorizing.

```python
Q1 = df['Sleep Duration'].quantile(0.25)
Q3 = df['Sleep Duration'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['Sleep Duration'] >= lower_bound) & (df['Sleep Duration'] <= upper_bound)]

print(df_cleaned[['Sleep Duration']].describe())
```

karena sleep duration masih memiliki  outlier maka digunakan Interquartile Range (IQR) untuk mendeteksi dan menghapus outlier pada kolom Sleep Duration. Metode IQR adalah cara umum untuk mendeteksi outlier dengan mengukur jarak antara kuartil pertama (Q1) dan kuartil ketiga (Q3). Nilai-nilai yang berada di luar batas bawah dan batas atas (dihitung dari IQR) dianggap sebagai outlier.

```python
plt.boxplot(df_cleaned['Sleep Duration'])
plt.title('Sleep Duration (Cleaned)')
plt.show()
```

boxplot untuk variabel Sleep Duration yang sudah dibersihkan dari outlier menggunakan metode IQR.
